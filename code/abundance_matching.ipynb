```python
#load the observational data (positions only)
data_path = "../data/"
data_filename = "2MRSSGfogs_checked.dat"
data_obs = loadtxt(data_path+data_filename)
```

```python
#load the observational data (includes the log10 of the luminosities)
mags_filename = "TWOMRSp2.dat"
data_mags = loadtxt(data_path+mags_filename)
```

```python
# select all galaxies in a sphere
p_x = data_obs[:,0]
p_y = data_obs[:,1]
p_z = data_obs[:,2]
p_radius = sqrt(p_x**2 + p_y**2 + p_z**2)
radius = 50.0
in_radius = where(p_radius < radius)
in_radius = in_radius[0]
data_obs = data_obs[in_radius,:]
data_mags = data_mags[in_radius,:]
```

```python
obs_x = data_obs[:,0]
obs_y = data_obs[:,1]
obs_z = data_obs[:,2]
obs_lum = (40.0 - data_mags[:,3])/2.5 # is this conversion correct?
```

```python
#load the simulation data
data_path = "../data/"
data_filename = "2MRS_180Mpc_i1759_0768_snap992_all_z0.000.AHF_halos"
data_sim = loadtxt(data_path+data_filename)
```

```python
p_x = data_sim[:,5]
p_y = data_sim[:,6]
p_z = data_sim[:,7]
mass_data = data_sim[:,3]
scale = 1000.0 #kpc/h to Mpc
center = 90.0 # center of the bos in Mpc
p_x = (p_x/scale - center)
p_y = (p_y/scale - center)
p_z = (p_z/scale - center)
```

```python
p_radius = sqrt(p_x**2 + p_y**2 + p_z**2)
radius = 50.0
in_radius = where(p_radius < radius)
in_radius = in_radius[0]
data_sim = data_sim[in_radius,:]
sim_x = (data_sim[:,5]/scale - center)
sim_y = (data_sim[:,6]/scale - center)
sim_z = (data_sim[:,7]/scale - center)
sim_mass = data_sim[:,3]
```

```python
log_mass = log10(sim_mass)
hist, bins = histogram(log_mass, bins=15)
hist_lum, bins_lum = histogram(obs_lum, bins=15)

volume = pi*(4.0/3.0)*radius**3
plot(bins[1:], log10(hist/volume))
plot(bins_lum[1:], log10(hist_lum/volume))
```

```python
# group the data to make the k-means clustering
n_galaxies = size(obs_lum)
n_halos = size(sim_mass)
print n_galaxies, n_halos
```

```python
# create the IDs.
gal_ID = empty(n_galaxies)
gal_ID[:] = 0
halo_ID = empty(n_halos)
halo_ID[:] = 1

obj_ID = empty((0))
obj_ID = append(obj_ID, gal_ID)
obj_ID = append(obj_ID, halo_ID)
```

```python
# create a long list of positions
obj_x = empty((0))
obj_x = append(obj_x, obs_x)
obj_x = append(obj_x, sim_x)

obj_y = empty((0))
obj_y = append(obj_y, obs_y)
obj_y = append(obj_y, sim_y)

obj_z = empty((0))
obj_z = append(obj_z, obs_z)
obj_z = append(obj_z, sim_z)

```

```python
#create a long list of properties (luminosity, mass)
obj_prop = empty((0))
obj_prop = append(obj_prop, obs_lum)
obj_prop = append(obj_prop, log_mass)

print size(obj_x), size(obj_y), size(obj_z), size(obj_ID), size(obj_prop) 
```
