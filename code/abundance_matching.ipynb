```python
#load the simulation data
data_path = "../data/"
data_filename = "2MRS_180Mpc_i1759_0768_snap992_all_z0.000.AHF_halos"
data_sim = loadtxt(data_path+data_filename)
```

```python
# group the data to make the k-means clustering
n_galaxies = size(obs_lum)
n_halos = size(sim_mass)
print n_galaxies, n_halos
```

```python
# create the IDs.
gal_ID = empty(n_galaxies)
gal_ID[:] = 0
halo_ID = empty(n_halos)
halo_ID[:] = 1

obj_ID = empty((0))
obj_ID = append(obj_ID, gal_ID)
obj_ID = append(obj_ID, halo_ID)
```

```python
# create a long list of positions
obj_x = empty((0))
obj_x = append(obj_x, obs_x)
obj_x = append(obj_x, sim_x)

obj_y = empty((0))
obj_y = append(obj_y, obs_y)
obj_y = append(obj_y, sim_y)

obj_z = empty((0))
obj_z = append(obj_z, obs_z)
obj_z = append(obj_z, sim_z)
```

```python
#create a long list of properties (luminosity, vmax)
obj_prop = empty((0))
obj_prop = append(obj_prop, obs_lum)
obj_prop = append(obj_prop, sim_vmax)

obj_mass = empty((0))
obj_mass = append(obj_mass, obs_lum)
obj_mass = append(obj_mass, sim_mass)

print size(obj_x), size(obj_y), size(obj_z), size(obj_ID), size(obj_prop), size(obj_mass)
```

```python
n_points =size(obj_x)
features = empty((n_points,3))
features[:,0] = obj_x[:]
features[:,1] = obj_y[:]
features[:,2] = obj_z[:]
```

```python
n_clusters = 1
centroids,_ = kmeans(features,n_clusters, iter=1)
idx,_ = vq(features,centroids)
```

```python
sizes = empty((0))
n_gals = empty((0))
n_halos = empty((0))
gal_lum_all = empty((0))
halo_vmax_all = empty((0))
halo_mass_all = empty((0))
group_id = arange(n_clusters)
for i in range(n_clusters):
    #select the halos and galaxies in this gropu
    id_group_gals = where((idx==i) & (obj_ID==0))
    id_group_halos = where((idx==i) & (obj_ID==1))
    
    #count their total numbers
    n_gals_group = size(id_group_gals)
    n_halos_group = size(id_group_halos)
    n_gals = append(n_gals, n_gals_group)
    n_halos = append(n_halos, n_halos_group)
    if ((n_gals_group>0) & (n_gals_group < n_halos_group)):
        
   # if n_halos>n_gals:
        #get their luminosities and masses
        gal_lum_group = obj_prop[id_group_gals]
        halo_vmax_group = obj_prop[id_group_halos]
        halo_mass_group = obj_mass[id_group_halos]
        
        #get them sorted
        halo_vmax_sorted = sort(halo_vmax_group)
        halo_vmax_sorted_id = argsort(halo_vmax_group)
        halo_mass_sorted = halo_mass_group[halo_vmax_sorted_id[-n_gals_group:]]
        halo_vmax_selected = halo_vmax_sorted[-n_gals_group:]
        gal_lum_sorted = sort(gal_lum_group)
    
        #append them to a single array
        gal_lum_all = append(gal_lum_all, gal_lum_sorted)
        halo_vmax_all = append(halo_vmax_all, halo_vmax_selected)
        halo_mass_all = append(halo_mass_all, halo_mass_sorted)
print size(gal_lum_all), size(halo_vmax_all), size(halo_mass_all)
```
